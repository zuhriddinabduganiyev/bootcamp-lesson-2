#### 2.16 Bu kod df DataFrameida "test preparation course" ustuni "completed" qiymati bilan va "gender" ustuni "male" (erkaklar) qiymati bilan solishtirib, bu shartlarga mos keladigan qatorlarni olib beradi. Ya'ni tayyorlangan va erkak bo'lgan talabalar haqida ma'lumotni chiqaradi.

```python
is_completed = df["test preparation course"] == "completed"
is_male = df["gender"] == "male"
df[is_completed & is_male]
```

#### 2.16 Quyidagi kod df DataFrameida "test preparation course" ustuni "none"  qiymati bilan va "gender" ustuni "male" (erkaklar) qiymati bilan solishtirib, bu shartlarga mos keladigan qatorlarni olib beradi. Ya'ni, tayyorlanmagan va erkak bo'lgan talabalar haqida ma'lumotni chiqaradi.

```python
is_completed = df["test preparation course"] == "none"
is_male = df["gender"] == "male"
df[is_completed & is_male]
```

#### 2.17 Quyidagi kod df DataFrameida "test preparation course" ustuni "completed" yoki "none" qiymatlari bilan solishtirib bu shartlarga mos keladigan qatorlarni olib beradi ya'ni "completed" yoki "none" bo'lgan barcha talabalar haqida ma'lumotni chiqaradi.

```python
is_completed = df["test preparation course"].isin(["completed", "none"])
df[is_completed]
```
